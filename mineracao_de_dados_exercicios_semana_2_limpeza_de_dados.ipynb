```python
import pandas as pd
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/higoramario/univesp-com360-mineracao-dados/main/dados-covid-limpeza.csv'
casos_covid = pd.read_csv(url)
casos_covid.head(20)
```

```python
casos_covid.info()
```

```python
casos_covid.describe()
```

Apesar do atributo "vacina" estar em formato numérico ele é uma informação binária (tomou ou não tomou vacina) logo apenas a informação "count" faz sentido.

```python
casos_covid['uf'].value_counts()
```

Vemos aqui que existem registros diferentes para o mesmo estado ("SP" e "sp", por exemplo), logo é necessário corrigir esses problemas.
Como os dados são apenas textos, podemos utilizar a função .str.upper() para padronizar todas as entradas em maiúsculo.

```python
casos_covid['uf'] = casos_covid['uf'].str.upper()
casos_covid['uf'].value_counts()
```

```python
casos_covid['renda'].value_counts()
```

```python
casos_covid['vacina'].value_counts()
```

Preenchimento dos dados vazios da coluna idade com o valor da média arredondada das idades coletadas.

```python
casos_covid['idade'].fillna(round(casos_covid['idade'].mean()), inplace=True)
casos_covid['idade'].value_counts()
```

Preenchimento dos dados vazios da coluna UF com o último valor observado através do parâmetro method=ffill.

```python
casos_covid['uf'].fillna(method='ffill', inplace=True)
casos_covid['uf'].value_counts()
```

Preenchimento dos dados vazios da coluna Renda com o próximo valor observado através do parâmetro method=bfill.

```python
casos_covid['renda'].fillna(method='bfill', inplace=True)
casos_covid['renda'].value_counts()
```
